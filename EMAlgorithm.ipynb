# An Introduction to the EM Algorithm and Finite Mixture Models

To thoroughly understand the EM algorithm and its use in estimating the parameters of finite mixture models, a prerequisite is a foundation in convex analysis, useful also in many well-behaved optimization problems. We begin with a definition, followed by a simple lemma.

## Convex Analysis

### Definition 1

Let $I = (a, b) \subseteq \mathbb{R}$ and let $f: I \rightarrow \mathbb{R}$. We say $f$ is *convex* on $I$ if 

$$f(\lambda x_1 + (1 - \lambda) x_2) \leq \lambda f(x_1) + (1 - \lambda) f(x_2)$$

for every $x_1, x_2 \in I$ and $\lambda \in [0, 1]$. We say $f$ is *strictly convex* on $I$ if the inequality is strict for all $\lambda \in (0, 1)$.

Geometrically, this says that if $f$ is convex on $I$, then for any $x, y \in I$, the line segment from $(x, f(x))$ to $(y, f(y))$ will always stay 'above' the function $f$:

```python
def f(x):
    return (x - 4)**2

x = np.linspace(0, 8, 1000)
y = f(x)
fig = plt.figure(figsize=(10, 8))
plt.plot(x, y)
plt.plot([2, 7], [f(2), f(7)], 'r')
```

### Fact 1

$-\ln(x)$ is convex on the interval $(0, \inf)$.

#### Proof (by picture, for sake of time)

```python
def f(x):
    return -np.log(x)

x = np.linspace(0.01, 10, 1000)
y = f(x)
fig = plt.figure(figsize=(10, 8))
plt.plot(x, y)
plt.plot([1, 8], [f(1), f(8)], 'r')
```

We now give one of the most important inequalities of all of mathematics: Jensen's Inequality (at least a subcase of it). Many other important inequalities are simple cases of Jensen's Inequality, including Young's, Holder's, Minkowski's, and the Arithmetic-Geometric Mean inequalities.

### Theorem 1 (Jensen's Inequality)

Let $f$ be a convex function defined on an interval $I \in \mathbb{R}$. Let $x_1, \cdots, x_n \in I$ and $\lambda_1, \cdots, \lambda_n \geq 0$ such that $\sum_{i=1}^n \lambda_i = 1$. Then

$$f\left(\sum_{i=1}^n \lambda_i x_i \right) \leq \sum_{i=1}^n \lambda_i f(x_i).$$

#### Proof

Let $n = 1$. Then we have $f(x_1) \leq f(x_1)$ trivially. If $n = 2$, then this is the definition of a convex function. Suppose the inequality holds for some $n$. Then 

\begin{align}
f\left(\sum_{i=1}^{n + 1} \lambda_i x_i\right) &= f\left(\sum_{i=1}^n \lambda_i x_i + \lambda_{n+1} x_{n+1}\right) \\
&= f\left((1 - \lambda_{n+1})\frac{1}{1 - \lambda_{n+1}} \sum_{i=1}^n \lambda_i x_i + \lambda_{n+1}x_{n+1}\right) \\
&\leq (1 - \lambda_{n+1}) f\left(\frac{1}{1 - \lambda_{n+1}} \sum_{i=1}^n \lambda_i x_i \right) + \lambda_{n+1} f(x_{n+1}) \\
&= (1 - \lambda_{n+1}) f\left( \sum_{i=1}^n \frac{\lambda_{i}}{1 - \lambda_{n+1}} x_i \right) + \lambda_{n+1}f(x_{n+1}) \\
&\leq (1 - \lambda_{n+1})\sum_{i=1}^n \frac{\lambda_i}{1 - \lambda_{n+1}} f(x_i) + \lambda_{n+1}f(x_{n+1}) \\
&= \sum_{i=1}^{n+1} \lambda_i f(x_{i})
\end{align}

### Corollary 1

Let $x_1, \cdots, x_n > 0$ and let $\lambda_1, \cdots, \lambda_n \geq 0$ such that $\sum_{i=1}^n \lambda_i = 1.$ Then 

$$\ln\left( \sum_{i=1}^n \lambda_i x_i \right) \geq \sum_{i=1}^n \lambda_i \ln x_i.$$

### Theorem 2 (without proof)

a) If $f_1, \cdots, f_k$ are convex functions over a convex set $C \subseteq \mathbb{R}^n$ then

$$g(\mathbf{x}) = \sum_{i=1}^k f_i(\mathbf{x})$$

is also convex.

b) If $f$ is a convex function over a convex set $C \subseteq \mathbb{R}^n$, and if $c \in \mathbb{R}$ is nonnegative, then $cf$ is also convex.

c) Linear functions are convex.

## Expectation Maximization

The EM algorithm is a method used to compute the maximum likelihood estimates of parameters in a statistical model in which there are latent (hidden) variables.

In general, given a statistical model with known data $O$ and unobserved data $X$ (with a one-to-one correspondence), unknown parameters $\lambda$, and a likelihood function $L(\lambda; X, O) = \mathbb{P}(O, X | \lambda)$ the objective is to find $\lambda^{*}$ which maximizes $L(\lambda; O)$. Note that we can marginalize out the unobserved data by noting that

$$L(\lambda; O) = \mathbb{P}(O | \lambda) = \sum_{X} \mathbb{P}(O, X | \lambda)$$

though this computation is often untractable. Instead, we follow an iterative procedure as follows:

(i) Given $\lambda_t$, compute $Q(\lambda | \lambda_t) = E_{X | O, \lambda_t} [\ln \mathbb{P}(O, X | \lambda)] = \sum_X \mathbb{P}(X | O, \lambda_t) \ln \mathbb{P}(O, X | \lambda).$

(ii) Update $\lambda_{t+1} = \arg\max_{\lambda} Q(\lambda | \lambda_t)$.

You should be asking yourself the following question: is $\mathbb{P}(O | \lambda_{t+1}) \geq \mathbb{P}(O | \lambda_t)$ necessarily true? Otherwise, why would this procedure help?

The key lies in our results of convex analysis and Jensen's inequality. First, since $\ln$ is a strictly increasing function, then the parameter value $\lambda^{*}$ which maximizes $l(\lambda) = \ln \mathbb{P}(O | \lambda)$ must also maximize $\mathbb{P}(O | \lambda)$, and so we deal here with the log-likelihood function $l(\lambda)$.

### Lemma 1

For any set of parameters $\lambda$ and current estimate $\lambda_t$,

$$l(\lambda) \geq l(\lambda_t) + \sum_X \mathbb{P}(X | O, \lambda_t) \ln \frac{\mathbb{P}(O, X | \lambda)}{\mathbb{P}(O, X | \lambda_t)}.$$

#### Proof

Note that

\begin{align}
l(\lambda) - l(\lambda_t) &= \ln \mathbb{P}(O | \lambda) - \ln \mathbb{P}(O | \lambda_t) \\
&= \ln\left(\sum_X \mathbb{P}(O, X | \lambda)\right) - \ln \mathbb{P}(O | \lambda_t) \\
&= \ln \left(\sum_X \mathbb{P}(X | O, \lambda_t) \frac{\mathbb{P}(O, X | \lambda)}{\mathbb{P}(X | O, \lambda_t)} \right) - \ln \mathbb{P}(O | \lambda_t) \\
&\geq \sum_X \mathbb{P}(X | O, \lambda_t) \ln \frac{\mathbb{P}(O, X | \lambda)}{\mathbb{P}(X | O, \lambda_t)} - \ln \mathbb{P}(O | \lambda_t) \\
&= \sum_X \mathbb{P}(X | O, \lambda_t) \ln \frac{\mathbb{P}(O, X | \lambda)}{\mathbb{P}(X | O, \lambda_t) \mathbb{P}(O | \lambda_t)} \\
&= \sum_X \mathbb{P}(X | O, \lambda_t) \ln \frac{\mathbb{P}(O, X | \lambda)}{\mathbb{P}(O, X | \lambda_t)}
\end{align}

Now we can answer that question we had about EM:

### Theorem 2

If $\lambda_t$ and $\lambda_{t+1}$ are parameter estimates for consecutive iterations of the EM algorithm, then $\mathbb{P}(O | \lambda_{t+1}) \geq \mathbb{P}(O | \lambda_t).$

#### Proof

Note that

\begin{align}
\lambda_{t+1} &= \arg\max_{\lambda} \sum_X \mathbb{P}(X | O, \lambda_t) \ln \mathbb{P}(O, X | \lambda) \\
&= \arg\max_{\lambda} \sum_X \mathbb{P}(X | O, \lambda_t) \ln \mathbb{P}(O, X | \lambda) - \sum_X \mathbb{P}(X | O, \lambda_t) \ln \mathbb{P}(O, X | \lambda_t) \\
&= \arg \max_{\lambda} \sum_X \mathbb{P}(X | O, \lambda_t) \ln \frac{\mathbb{P}(O, X | \lambda)}{\mathbb{P}(O, X | \lambda_t)}.
\end{align}

By **Lemma 1** and the above computation, we have

\begin{align}
l(\lambda_{t+1}) &\geq l(\lambda_t) + \sum_X \mathbb{P}(X | O, \lambda_t) \ln \frac{\mathbb{P}(O, X | \lambda_{t+1})}{\mathbb{P}(O, X | \lambda_t)} \\
&\geq l(\lambda_t) + \sum_X \mathbb{P}(X | O, \lambda_t) \ln \frac{\mathbb{P}(O, X | \lambda_t)}{\mathbb{P}(O, X | \lambda_t)} \\
&= l(\lambda_t) + \sum_X \mathbb{P}(X | O, \lambda_t) \ln 1 \\
&= l(\lambda_t).
\end{align}

Hmmm... that's clearly a very long right-tailed distribution. Let's look at it on the log-scale:

```python
_ = plt.hist(np.log(np.array(data) + 1), bins=50)
```

If you're like me, you'd see this and exclaim, "Holy Canoli! What is that???". My thoughts exactly when I saw just such a distribution for the first time. This clearly doesn't come from a normal, or an exponential. How might we model it, you ask?

## Finite Mixture Models to the Rescue!

We will hypothesize that this data is generated from a finite mixture model composed of three parts: an exponential (to get the stuff near 0), a point-mass (to get that weird spike) and a log-normal. What this means is that we suppose the data was generated as follows:

\begin{align}
x_i &\leftarrow \text{Cat}(\mathbf{w}) \\
o_i &\leftarrow f_{x_i}
\end{align}

where $f_{x_i}$ is the distribution for component $x_i$.

In our case, $f_0$ could be an exponential, $f_1$ a point-mass, and $f_2$ a log-normal. If we knew the corresponding $x_i$ for each value $o_i$, this would be easy. Of course, we don't (nor do the $x_i$'s even exist in reality), which is why the EM algorithm works so well (remember how it assumes some kind of latent variable upon which the observed variables are conditioned?).

Well, to use EM, it's a simple as computing

$$Q(\lambda | \lambda_t) = \sum_X \mathbb{P}(X | O, \lambda_t) \ln \mathbb{P}(O, X | \lambda)$$

and finding the argmax, right? Well, that would often be impossible (or at least a colossal waste of computing power). We can often simplify it down to our special case (and once you do this a few times, you can instinctually do this without the need of the mathematical rigor, though it never hurts).

In our case, it boils down to the following calculations:

For each observed value $o_i$ and component $k$, calculate 

\begin{align}
\gamma_{i, k} &= \mathbb{P}(k | o_i) \\
&\propto \mathbb{P}(o_i | k) \mathbb{P}(k) \\
&= \mathbb{P}(o_i | k) w_k
\end{align}

Then we can update the exponential parameter with the following update:

$$\widehat{\lambda_{exp}} = \frac{\sum_{i=1}^N \gamma_{i, 1} o_i}{\sum_{i=1}^N \gamma_{i, 1}}.$$

The mean of the log-normal gets updated as follows:

$$\widehat{\mu} = \frac{\sum_{i=1}^N \gamma_{i, 2} \ln(o_i)}{\sum_{i=1}^N \gamma_{i, 2}}.$$

The variance gets updated as follows:

$$\widehat{\sigma} = \frac{\sum_{i=1}^N \gamma_{i, 2} (\ln(o_i) - \hat{\mu})^2}{\sum_{i=1}^N \gamma_{i, 2}}.$$

The weights get updates as follows:

$$\widehat{w_k} = \frac{\sum_{i=1}^N \gamma_{i, k}}{N}.$$

We don't need to update any parameter for the point-mass (other than the weight, which we just did).

To see this in action, we'll use a custom class to learn the mixture model parameters to fit the above data, where we have specifically identified the point-mass to be at 30:

```python
model = LnExponPMMixture(pms=[30])
model.fit(np.array(data))
```

```python
comps, draws = model.sample(10000)
_ = plt.hist(np.log(np.array(draws) + 1), 50)
```
